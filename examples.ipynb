# Load Images from Local Storage

```python
# Read a PNG and resize it with padding
example_gif = ImageArtificer.from_local_disk("example_images/sunset.png")
example_gif.resize_and_pad_image((300,300))
```

```python
# Read a GIF and resize it with padding
example_gif = ImageArtificer.from_local_disk("example_images/golf_course.gif")
example_gif.resize_and_pad_image((300,300))
```

```python
# Read a JPG and resize it with padding
example2 = ImageArtificer.from_local_disk("example_images/example2.jpg")
example2.resize_and_pad_image(size=(300,300), padding_color='complementary')
```

```python
# Read a WEBP and resize it with padding
example2 = ImageArtificer.from_local_disk("example_images/golf_course.webp")
example2.resize_and_pad_image(size=(300,300))
```

```python
# Read a heic image (very slow)
# This would need pyheif, which requires istalling the libheif program on the machine - so I didn't mess with it.
heic_example = ImageArtificer.from_local_disk("example_images/tall_example.heic")
heic_example.resize_and_pad_image((500,500))
```

# Load an image from a URL on the web

```python
cat = ImageArtificer.from_url("https://d1hjkbq40fs2x4.cloudfront.net/2016-07-16/files/cat-sample_1313.jpg")
cat
```

```python
# Resize this instance of the image inplace
cat.resize_and_pad_image((300,300),inplace=True)
```

```python
# Save the file locally 
cat.save_to_local("cat.jpg")
```

# Search for an image on Google

```python
tacobell = ImageArtificer.from_google_search("Taco Bell Logo")
tacobell
```

```python
# Resize our taco bell log to a square and add a giltter overlay!
tacobell.resize_and_pad_image((400,400)).apply_overlay("overlays/glitter_top_bottom.png")
```

```python
blockbuster = ImageArtificer.from_google_search("blockbuster logo").resize_and_pad_image(size=(800,800))
blockbuster
```

The blockbuster logo centered in a 800x800 square image with a red ribbon and glitter overlay

```python
blockbuster.apply_overlay("overlays/ribbon_red.png").apply_overlay("overlays/glitter_top_bottom.png")
```

# Easily save to google cloud storage

```python
from google.cloud import storage
storage_client = storage.Client() # Default storage client read from the enviornment variables
bucket_name = 'bucket_name'
blockbuster.save_to_gcs(storage_client, bucket_name, "img_test/0")
```
